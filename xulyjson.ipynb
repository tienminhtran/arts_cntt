```python
import pandas as pd
import json

# Đọc file Excel
df = pd.read_excel("dangkithamdu.xlsx")

# Chuẩn hóa tên cột (xóa khoảng trắng thừa nếu có)
df.columns = df.columns.str.strip()

# Chuyển đổi kiểu dữ liệu + thêm id
def convert_row(row, idx):
    return {
        "id": str(idx + 1),  # MockAPI cần id là string
        "email": str(row["email"]) if pd.notna(row["email"]) else "",
        "mssv": int(row["mssv"]) if pd.notna(row["mssv"]) else 0,
        "lop": str(row["lop"]) if pd.notna(row["lop"]) else "",
        "sodienthoai": str(row["sodienthoai"]) if pd.notna(row["sodienthoai"]) else "",
        "khoa": str(row["khoa"]) if pd.notna(row["khoa"]) else "",
        "theloai": str(row["theloai"]) if pd.notna(row["theloai"]) else "",
        "hoten": str(row["hoten"]) if pd.notna(row["hoten"]) else "",
        "thoigianduthi": str(row["thoigianduthi"]) if pd.notna(row["thoigianduthi"]) else "",
        "khunggio": str(row["khunggio"]) if pd.notna(row["khunggio"]) else "",
        "tentietmuc": str(row["tentietmuc"]) if pd.notna(row["tentietmuc"]) else "",
        "thongtinbandonghanh": str(row["thongtinbandonghanh"]) if pd.notna(row["thongtinbandonghanh"]) else "",
        "trangthai": int(row["trangthai"]) if pd.notna(row["trangthai"]) else 0
    }

# Áp dụng cho từng dòng, thêm id tự động
records = [convert_row(row, idx) for idx, row in df.iterrows()]

# Xuất JSON
api_json = json.dumps(records, ensure_ascii=False, indent=2)

# Lưu ra file
with open("dangky.json", "w", encoding="utf-8") as f:
    f.write(api_json)

print(api_json)
```
